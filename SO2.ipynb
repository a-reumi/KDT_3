```python
# 숫자로 변환
co2DF = so2DF.apply(pd.to_numeric, errors='coerce')

# 데이터를 행렬 변환 (월을 X축, 도시를 Y축 데이터로 사용)
co2DF_T = co2DF.T  # 행과 열을 변환
co2DF_T.columns = so2DF.index  # 열 이름을 도시 이름으로 변경

# 연도별 평균 계산 (각 연도의 월별 데이터를 평균으로 집계)
yearly_avg = co2DF_T.groupby(co2DF_T.index.str[:4]).mean()

# 범례에서 숫자로 나타나는 부분을 도시 이름으로 변경
yearly_avg.columns = co2DF.index

# 2017년도 이후 데이터 필터링
yearly_avg_filtered = yearly_avg.loc['2017':'2023']

# 도시별 색상 맵핑
colors = {
    '총계':'#FF0000',  # 빨강
    '서울특별시':'#FF7F00',  # 주황
    '부산광역시':'#FFFF00',  # 노랑
    '대구광역시':'#008000',  # 초록
    '인천광역시':'#0000FF',  # 파랑
    '광주광역시':'#000080',  # 남색
    '대전광역시':'#800080',  # 보라
    '울산광역시':'#000000',  # 검정
    '세종특별자치시':'#FFC0CB',  # 연핑크
    '경기도':'#808080',  # 회색
    '강원특별자치도':'#ADFF2F',  # 연두
    '충청북도':'#87CEEB',  # 하늘
    '충청남도':'#D8BFD8',  # 연보라
    '전북특별자치도':'#6B8E23',  # 카키
    '전라남도':'#008B8B',   # 청록
    '경상북도':'#FF1493',  # 네온 핑크
    '경상남도':'#00FFBF',  # 밝은 민트
    '제주특별자치도':'#8B4513'   # 진한 갈색
}

# 꺾은선 그래프 생성
plt.figure(figsize=(20, 10))

for city in yearly_avg_filtered.columns:
    sns.lineplot(
        x=yearly_avg_filtered.index,
        y=yearly_avg_filtered[city],
        label=city,
        color=colors.get(city, 'black'),  # 도시별 색상 맵핑 
        marker='o'
    )

plt.xlabel("연도", fontsize=14)
plt.ylabel("SO2 평균 농도", fontsize=14)
plt.title("2017년부터 2023년까지 연도별 도시별 아황산가스 평균 농도 변화", fontsize=20)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="도시", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.grid(True)
plt.show()[-'']


```
