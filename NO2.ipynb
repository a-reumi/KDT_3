```python
# 데이터 정리
no2DF.replace('-', 0, inplace=True)  # '-' 값을 0으로 변환
no2DF.rename(columns={'Unnamed: 0': '지역'}, inplace=True)  # 첫 번째 컬럼 이름 변경


# # 숫자로 변환
# no2DF = no2DF.apply(pd.to_numeric, errors='coerce')

# 데이터를 행렬 변환 (월을 X축, 도시를 Y축 데이터로 사용)
# no2DF_T = no2DF.T  # 행과 열을 변환
# no2DF_T.columns = no2DF.index  # 열 이름을 도시 이름으로 변경

# 연도별 평균 계산 (각 연도의 월별 데이터를 평균으로 집계)
yearly_avg = no2DF_T.groupby(no2DF_T.index.str[:4]).mean()
# print(yearly_avg)
yearly_avg.drop(index='지역',inplace=True)

# 범례에서 숫자로 나타나는 부분을 도시 이름으로 변경
yearly_avg.columns =['총계','서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시',
       '울산광역시', '세종특별자치시', '경기도', '강원특별자치도', '충청북도', '충청남도', '전북특별자치도',
       '전라남도', '경상북도', '경상남도', '제주특별자치도']
# X축: 연도, Y축: 도시별 평균 NO₂ 농도 변화
plt.figure(figsize=(20, 10))
sns.lineplot(data=yearly_avg, markers='o', dashes=False)
plt.xlabel("연도")
plt.ylabel("NO2 평균 농도")
plt.title("연도별 도시별 이산화질소 평균 농도 변화", fontsize=20)
plt.xticks(rotation=45)
plt.legend(title="도시", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True) 

plt.show() 
```

```python
# 숫자로 변환
co2DF = no2DF.apply(pd.to_numeric, errors='coerce')

# 데이터를 행렬 변환 (월을 X축, 도시를 Y축 데이터로 사용)
co2DF_T = co2DF.T  # 행과 열을 변환
co2DF_T.columns = no2DF.index  # 열 이름을 도시 이름으로 변경

# 연도별 평균 계산 (각 연도의 월별 데이터를 평균으로 집계)
yearly_avg = co2DF_T.groupby(co2DF_T.index.str[:4]).mean()

# 열 이름을 명시적으로 도시 이름으로 설정
yearly_avg.columns = [
    '총계', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시',
    '울산광역시', '세종특별자치시', '경기도', '강원특별자치도', '충청북도', '충청남도',
    '전북특별자치도', '전라남도', '경상북도', '경상남도', '제주특별자치도'
]

# 2017년도 이후 데이터 필터링
yearly_avg_filtered = yearly_avg.loc['2017':'2023']

# 도시별 색상 맵핑
colors = {
    '총계':'#FF0000',  # 빨강
    '서울특별시':'#FF7F00',  # 주황
    '부산광역시':'#FFFF00',  # 노랑
    '대구광역시':'#008000',  # 초록
    '인천광역시':'#0000FF',  # 파랑
    '광주광역시':'#000080',  # 남색
    '대전광역시':'#800080',  # 보라
    '울산광역시':'#000000',  # 검정
    '세종특별자치시':'#FFC0CB',  # 연핑크
    '경기도':'#808080',  # 회색
    '강원특별자치도':'#ADFF2F',  # 연두
    '충청북도':'#87CEEB',  # 하늘
    '충청남도':'#D8BFD8',  # 연보라
    '전북특별자치도':'#6B8E23',  # 카키
    '전라남도':'#008B8B',   # 청록
    '경상북도':'#FF1493',  # 네온 핑크
    '경상남도':'#00FFBF',  # 밝은 민트
    '제주특별자치도':'#8B4513'   # 진한 갈색
}

# 꺾은선 그래프 생성
plt.figure(figsize=(20, 10))

for city in yearly_avg_filtered.columns:
    sns.lineplot(
        x=yearly_avg_filtered.index,
        y=yearly_avg_filtered[city],
        label=city,
        color=colors.get(city, 'black'),  # 도시별 색상 맵핑
        marker='o'
    )

plt.xlabel("연도", fontsize=14)
plt.ylabel("NO2 평균 농도", fontsize=14)
plt.title("연도별 도시별 이산화질소 평균 농도 변화", fontsize=20)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="도시", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.grid(True)
plt.show()
```
