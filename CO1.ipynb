```python
# 데이터 정리
co1DF.replace('-', 0, inplace=True)  # '-' 값을 0으로 변환
co1DF.rename(columns={'Unnamed: 0': '지역'}, inplace=True)  # 첫 번째 컬럼 이름 변경
co1DF.set_index('지역', inplace=True)  # '지역'을 인덱스로 설정

# 숫자로 변환
co2DF = co1DF.apply(pd.to_numeric, errors='coerce')

# 데이터를 행렬 변환 (월을 X축, 도시를 Y축 데이터로 사용)
co2DF_T = co2DF.T  # 행과 열을 변환
co2DF_T.columns = co2DF.index  # 열 이름을 도시 이름으로 변경

# 연도별 평균 계산 (각 연도의 월별 데이터를 평균으로 집계)
yearly_avg = co2DF_T.groupby(co2DF_T.index.str[:4]).mean()

# 범례에서 숫자로 나타나는 부분을 도시 이름으로 변경
yearly_avg.columns = co2DF.index

# X축: 연도, Y축: 도시별 평균 SO₂ 농도 변화
plt.figure(figsize=(20, 10))
sns.lineplot(data=yearly_avg, markers='o', dashes=False)
plt.xlabel("연도")
plt.ylabel("CO1 평균 농도")
plt.title("연도별 도시별 일산화탄소 평균 농도 변화", fontsize=20)
plt.xticks(rotation=45)
plt.legend(title="도시", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

plt.show()
```
